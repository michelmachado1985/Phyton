Removendo linhas duplicadas


```python

import pandas as pd
```

```python
#Abrindo o arquivo em excel e colocando na variavel arquivo.
arquivo = "C:\\Users\\miche\\Desktop\\Codigos Python\Datasets\\Aluno.xlsx"

#Carregando o arquivo
xls = pd.ExcelFile(arquivo)

#Verificando as planilhas
print(xls.sheet_names)
```

```python
alunos = xls.parse('Plan1')

#fazendo a leitura do arquivo.
alunos
```

```python
alunos.duplicated().head(10)
```

```python
alunos.Aluno.duplicated().head(10)
```

```python
alunos.duplicated().sum()
```

```python
#Para encontra a primeira ocorrência de repetição, começando do index 0.
alunos.loc[alunos.duplicated(keep='first'),:]
```

```python
#Para encontra a primeira ocorrência de repetição, começando do final do datafreme.
alunos.loc[alunos.duplicated(keep='last'),:]
```

```python

alunos.loc[alunos.duplicated(keep='first'),:]
```

```python

alunos.drop_duplicates(keep='first')
#Caso queira excluir no dataframe original se não só salvar o resultado do acima em uma nova variável.
#alunos.drop_duplicates(keep='first', inplace=True)
```
